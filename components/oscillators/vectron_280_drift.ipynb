```python
import sys
import numpy as np
import matplotlib.pyplot as plt
import time
from scipy.signal import find_peaks
import scipy.io as sio
%matplotlib inline
plt.rcParams['figure.figsize'] = [15, 10]


sys.path.append("gpib_instrument_control/")

import gpib_instrument_control.hp_5340a
counter=gpib_instrument_control.hp_5340a.Hp5340A()
counter.setResolution(0)
```

```python
print(counter.readValue())
```

```python
#matplotlib widget
fig = plt.figure()

hfig = display(fig, display_id=True)

t0=time.time()
xax=[]
yax=[]
counter.readValue()
while True:
    yax.append(1e6* ((counter.readValue()+200e6)/280e6))
    xax.append(time.time()-t0)
    
    
    plt.clf()
    plt.grid(True)
    plt.plot(xax, yax, 'b')
    fig.canvas.draw()
    hfig.update(fig)
    
    
    
```
